```python
from sympy import *
```

```python
s, z, T, d = symbols('s z T d')
k, kc, tau = symbols('K K_c \\tau')
```

```python
Gs = (kc * k) / (tau*s + 1)

Gz = Gs.subs(s, (2/T)*(z-1)/(z+1))
Hz = z**(-d/T)

Mz = Gz / (1 + Gz*Hz)

Mz
```

```python
j = simplify(Mz)

print(j)
```

```python
den = k*kc*T*(z + 1) + z**(d/T)*(T*(z + 1) + 2*tau*(z - 1))

vals = {
    d : 1*T,
    k : 1,
    kc : 1.189,
    tau : 0.01842499999999969,
}

den = den.subs(vals).subs(T, 0.3)

poles = roots(den, z)

poles = list(map(lambda x: abs(x.evalf()), poles))

poles
```
