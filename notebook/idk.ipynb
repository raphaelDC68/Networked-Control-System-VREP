```python
%matplotlib inline

from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Trc(Kc, G, tau):
    return tau * acos(-1/(Kc*G))/sqrt((Kc*G)**2 - 1)
```

```python
taus = [0.01842499999999969, 0.052075000000000315, 0.04757500000000192, 0.029274999999999718, 0.018287500000002677, 0.018975000000001074]

Kcs = np.linspace(1.01, 100, num=1000)
Trcs = []

for Kc in Kcs:
    chs = list(map(lambda x: Trc(Kc, 1, x), taus))
    Trcs += [chs]

i = -1
for ch in np.asarray(Trcs).transpose():
    print(i := i + 1, max(*ch))
    plt.plot(Kcs, ch)
plt.show()
```

```python
lims = np.linspace(0.35, 0, endpoint=False, num=3)
mid = lims - (lims[0] - lims[1])/2

for i in range(3):
    print(f'region {i}: {lims[i]}')  
    for Trc, Kc in zip(Trcs, Kcs): 
        if Trc[1] <= mid[i]:
            print(f'Kc: {Kc}')
            break
```
